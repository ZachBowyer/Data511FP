```python
import pandas as pd
import numpy as np
import os
```

# Read in datasets

```python
production = pd.read_csv('../Datasets/Production_Crops_Livestock_E_All_Data_(Normalized).csv', encoding='latin-1')
emissionCounts = pd.read_csv('../Datasets/Food_Production_Emissions.csv')
```

# Drop unnecessary columns

```python
emissionCounts = emissionCounts[['Food product', 'Total_emissions']]
production = production[['Area', 'Year', 'Item', 'Unit', 'Value']]
emissionCounts = emissionCounts.rename(columns={"Food product": "Item", "Total_emissions": "emissionperKGProduct"})
production = production.loc[production['Unit'] == 'tonnes']
```

```python
print(production['Item'].unique(), "\n\n", emissionCounts['Item'].unique())
production = production.loc[production['Item'].isin(['Wheat', 'Maize (corn)', 'Barley', 'Rice', 'Potatoes', 'Cassava, fresh', 'Sugar cane', 'Peas, green', 'Treenuts, Total', 'Groundnuts, excluding shelled', 'Palm oil', 'Sunflower-seed oil, crude', 'Rapeseed or canola oil, crude', 'Olive oil', 'Tomatoes', 'Leeks and other alliaceous vegetables', 'Citrus Fruit, Total', 'Bananas', 'Apples', 'Grapes', 'Wine', 'Coffee, green', 'Beef and Buffalo Meat', 'Meat of sheep', 'Meat of pig with the bone, fresh or chilled', 'Milk, Total', 'Cheese (All Kinds)', 'Eggs Primary', 'Meat of chickens, fresh or chilled', 'Meat of goat, fresh or chilled'])]
```

```python
#production
#emissionCounts #(Emissions are in kg of c02 equivalents)
```

```python
emissionCounts = emissionCounts.loc[emissionCounts['Item'].isin(['Apples', 'Barley (Beer)', 'Berries & Grapes', 'Maize (Meal)', 'Olive Oil', 'Potatoes', 'Rice', 'Cane Sugar', 'Sunflower Oil', 'Wheat & Rye (Bread)', 'Beef (dairy herd)', 'Cheese', 'Citrus Fruit', 'Eggs', 'Milk', 'Nuts', 'Onions & Leeks', 'Pig Meat', 'Palm Oil', 'Peas', 'Tomatoes', 'Wine', 'Bananas', 'Groundnuts', 'Rapeseed Oil', 'Cassava', 'Coffee', 'Poultry Meat', 'Goat'])]
print(production['Item'].unique())
print("\n\n", emissionCounts['Item'].unique())
#print(production['Area'].unique())
```

```python
production = production.replace('Maize (corn)', 'Corn')
production = production.replace('Sunflower-seed oil, crude', 'Sunflower oil')
production = production.replace('Beef and Buffalo Meat', 'Beef')
production = production.replace('Cheese (All Kinds)', 'Cheese')
production = production.replace('Citrus Fruit, Total', 'Citrus Fruit')
production = production.replace('Eggs Primary', 'Eggs')
production = production.replace('Milk, Total', 'Milk')
production = production.replace('Treenuts, Total', 'Treenuts')
production = production.replace('Leeks and other alliaceous vegetables', 'Leeks')
production = production.replace('Meat of pig with the bone, fresh or chilled', 'Pig')
production = production.replace('Peas, green', 'Peas')
production = production.replace('Groundnuts, excluding shelled', 'Groundnuts')
production = production.replace('Rapeseed or canola oil, crude', 'Rapeseed Oil')
production = production.replace('Cassava, fresh', 'Cassava')
production = production.replace('Coffee, green', 'Coffee')
production = production.replace('Olive oil', 'Olive Oil')
production = production.replace('Palm oil', 'Palm Oil')
production = production.replace('Sugar cane', 'Cane Sugar')
production = production.replace('Sunflower oil', 'Sunflower Oil')
production = production.replace('Meat of chickens, fresh or chilled', 'Chicken')
production = production.replace('Meat of goat, fresh or chilled', 'Goat')

print("\n\n", emissionCounts['Item'].unique())
```

```python
print(production['Item'].unique())
print("\n\n", emissionCounts['Item'].unique())
```

```python
emissionCounts = emissionCounts.replace('Barley (Beer)', 'Barley')
emissionCounts = emissionCounts.replace('Berries & Grapes', 'Grapes')
emissionCounts = emissionCounts.replace('Maize (Meal)', 'Corn')
emissionCounts = emissionCounts.replace('Wheat & Rye (Bread)', 'Wheat')
emissionCounts = emissionCounts.replace('Beef (dairy herd)', 'Beef')
emissionCounts = emissionCounts.replace('Pig Meat', 'Pig')
emissionCounts = emissionCounts.replace('Onions & Leeks', 'Leeks')
emissionCounts = emissionCounts.replace('Nuts', 'Treenuts')
emissionCounts = emissionCounts.replace('Poultry Meat', 'Chicken')
```

```python
print(production['Item'].unique())
print("\n\n", emissionCounts['Item'].unique())
```

```python
#production
#emissionCounts #(Emissions are in kg of c02 equivalents)
```

```python
exp = pd.merge(production, emissionCounts, on='Item', how='inner')
```

```python
print(production['Item'].unique())
#Greenhouse emmisions(Kg CO2 - equivalents per kg product) 
#So we know KG CO2 equivalents per kg of product, and we know metric tons of products. So to get total emisssions we convert mtons to kg and multiply.
#Value * 1000 * Total_emissions = KG OF C02 emissions of a product that year
exp['productionKG'] = exp.Value * 1000
exp['totalCO2'] = exp.productionKG * exp.emissionperKGProduct
exp
```

# Save df to new file 

```python
from pathlib import Path  
filepath = Path('../Datasets/FoodProductionTimeSeriesCO2Impact.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
exp.to_csv(filepath)  
```
